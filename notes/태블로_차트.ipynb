# 3) 범프차트
시간에 따른 순위변동을 확인하고싶다. 중분류 별 매출의 순위가 시간에 따라 어떻게 바뀌는지를 살펴보기 위해 열에 주문날짜를 얹고 (연속형), 매출 합의 순위를 꺾은선그래프로 나타낸다.

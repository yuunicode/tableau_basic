## 1) 집합 작업(동작)  
다이나믹하게 반응하는 대시보드를 만들기위해 필수적으로 쓰는 기능으로, **내가 변경해가며 보고싶은 필드** 를 하나의 **집합** 으로 만든 후, 그 멤버를 하나씩 선택하며 조회하는 원리다. 다음의 단계가 필요하다.  
  
(1) 값을 선택할 필드의 집합을 만든다.  
(2) 1단계의 집합을 변경하는 작업(동작)을 만든다.  
  
내가 한 프랜차이즈 대표라는 가정 하, 전국의 가맹점 매출들을 확인한다하면, 집합은 '각 시도 or 시군구'가 될 수 있다. 지역별로 인사이트를 얻을 수 있는 대시보드를 만들어보자.  
  
예를들어, 강원도를 선택하면 강원도의 전국 대비 매출 비중, 제품중분류별 전체 대비 매출 비중을 파란색으로 표시하고, 강원도와 경기도를 함께 선택하면 그 비중이 합쳐서 표시되도록 해보자.  
  
(1) 값을 선택할 필드의 집합을 만든다.  
시도 선택 집합이 필요하다. **시도선택집합**을 만든 후, 대시보드에 포함될 시트들에 시도선택집합을 색상으로 표시한다.    
집합은 IN/OUT 개념으로 작동하기 때문에, 내가 선택한 시도가 집합의 IN, 아닌 시도는 OUT이 되는 것이다. 색상에 집합을 추가하면, 전체는 회색으로 변한다.  

(2) 선택에 따라 집합 값이 변하도록 동작을 만든다. 
대시보드>동작> 집합 값 변경으로 선택  
  
   
  
원본시트는 대시보드에서 내가 원하는 시트, 대상집합은 데이터소스의 시도선택집합으로 선택한다. 또한 선택해지에서는 자기가 원하는 스타일대로 하면 된다.  
  
    
  
시도선택집합이 IN/OUT으로 설정돼있다. IN이 안쪽으로(선택한 값) 가길 원한다면 마크에있는 집합의 정렬을 수동으로 바꾸고 옮기면된다.  
  
  
  


## 2) 드릴다운
매개변수를 이용해서, 어떤 값을 선택했을 때 매개변수와 같다면 그 값을 변경해보는 실습을 해본다.  
제품이 대분류, 중분류가 있을 때, 전체일 때 대분류 수준의 매출을 살펴보고 특정 대분류 막대를 클릭했을 땐 해당 대분류의 중분류 수준으로 매출과 매출 구성 비율이 보이도록 해보자.  
  
  1. [제품대분류 매개변수] 생성  
  2. [제품중분류_드릴다운] 계산식 생성. 식은 다음과 같다.  
    
  IF [제품 대분류] = [제품대분류 매개변수] THEN [제품 중분류]
  ELSE [제품 대분류]
  END

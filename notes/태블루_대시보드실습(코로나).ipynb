# 실습과 서버 게시
원격으로 데이터를 업데이트할 수 있다. 다음 사진처럼 데이터추출을 하면 원통이 두개가 된다. 해당 데이터는 존스홉킨스 대학교에서 제공하는 코로나19 확진자 및 사망자 통계 중 일부 국가의 데이터를 발췌한 것이다.   
  

  
그럼 기존 대시보드를 더 발전시켜보자. 아래는 기존 대시보드다.  
  

  
시각적 분석의 첫번째는 질문으로 시작하는 것이다. 해당 질문에 맞는 차트 유형과 뷰를 만들고, 디자인을 하는 순서를 거친다. 질문을 다음과 같다고 하자.  
  1. 각 국가별 확진자 및 사망자 수는 얼마나될까? 어느 나라가 더 많을까?
  2. 각 대륙별 확진자 분포는 어떻고, 어떤 특징을 보일까?
  3. 각 국가별 확진자의 증가/감소폭을 확인하자.  
  
신병훈련소에서 제시하는 디렉션 말고도 내가 바꾸고싶은대로 자유롭게 바꿈


국가별 발생 건수를 조절 후 살펴보니 사망자가 확진자 대비 확연히 적어 대비가 잘 안되는 것을 확인할 수 있다.  
  
이 경우 '각 행 또는 열에 독립적인 축 범위'를 지정하면 된다. (축편집)  
그 후, 정렬하면 끝  
  
  
  

다음으로 지역별 확진자 분포를 확인할려면 어떻게 해야할까?


확진자가 전체인구수 대비 몇프로인지를 색감으로 확인하기 위해 마크를 맵으로 하면, 다음과 같다.  


다음으로는 일자별 확진자와 사망자수를 나타내는 그래프가 범위로 돼있는데, 라인으로 마크를 바꾼 후, 선택한 국가를 제외한 나머지를 회색처리를 해본다. (집합 이용)  
  
  
  
  \*Country Color -  선택된국가들 (집합에 속해있으면 색깔표시, 라벨표시)  
  \*Country Region - 지역
  
선택된 집합에 따라 선택되지않은 도시들은 회색으로 표시되도록 변경했다.  


다음으로는 대시보드를 손볼것이다. 국가별 발생 건수의 막대그래프에서 국가(Country Region)을 어떻게 선택하느냐에 따라서 다른 시트들에 영향이 가도록 한다.  


  
적절하게 Switcher 매개변수를 추가해서 New/Total Case를 고르도록 유도하고, 동작들을 추가해서 내가 선택한 나라에만 하이라이트가 나오도록 설정했다.
